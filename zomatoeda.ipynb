```python
df_country=pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head()
```

#Let us combine these 2 dataframes using countrycode

```python
final=pd.merge(df,df_country,on="Country Code",how='left')
```

```python
final.head(2)
```

```python
#to check data types
final.dtypes
```

```python
final.columns
```

```python
#checking records as per country
counrynames=final.Country.value_counts()
counrynames
#India has highest orders
```

```python
country=counrynames.index
country
```

```python
value=counrynames.values
value
```

```python
#Creating a pie chart:here we are looking at top 3 countries
plt.pie(value[:3],labels=country[:3],autopct='%1.2f%%')
```

Observation: India has highest zomato records worldwide followed by USA and UK.

```python
#there are columns Aggregate rating  which is of float64 type
#Rating color             object
#Rating text              object:let us group them up
rating=final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Ratings count'})
rating
```

Observation:When rating is between 4.5-4.9 it is Excellent
            When rating is between 4.0-4.4 it is Very Good
            When rating is between 3.5-3.9 it is Good
            When rating is between 2.5-3.4 it is Average
            When rating is between 0.0-2.4 it is Good

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Ratings count',data=rating)
```

```python
#we can colour the bars as per color provided in Ratings color

sns.barplot(x='Aggregate rating',y='Ratings count',hue='Rating color',data=rating)
#but we need to map the same as well because all of them do not match
```

#so we use palette
sns.barplot(x='Aggregate rating',y='Ratings count',hue='Rating color',data=rating,palette=['white','red','orange','yellow','green','dgreen'])

a.Not rated is most high
