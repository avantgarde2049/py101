## Feature engineering


```python
#we wish to Date_of_Journey to be split into day,month and year
#because currently it is a string oject needs to be in numeric
```

```python
df['day']=df['Date_of_Journey'].str.split('/').str[0]
df['day']
```

```python
df['month']=df['Date_of_Journey'].str.split('/').str[1]
df['month']
```

```python
df['year']=df['Date_of_Journey'].str.split('/').str[2]
df['year']
```

```python
df.head()
```

```python
#we can now drop the Date_of_Journey column
df.drop(['Date_of_Journey'],axis=1,inplace=True)
df.head()
```

```python
df.info()
```

we wish to typecast the new features created into int

```python
df['day']=df['day'].astype(int)
df['month']=df['month'].astype(int)
df['year']=df['year'].astype(int)
df.info()
```

```python
df.head()
```

```python
#now in Arrival_Time column we want the time common in all rows , so we split
df['Arrival_Time']=df['Arrival_Time'].str.split(' ').str[0]
```

```python
df.head()
```

```python
#searching for  null values
df.isnull().sum()
```

```python
#now we wish to split the Arrival_Time into hours and mins column

df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
```

```python

df['Arrival_min']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head(1)
```

```python
df.info()
```

```python
#typecast the new features into int

df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
df.info()
```

```python
#drop Arrival_Time column
df.drop('Arrival_Time',axis=1,inplace=True)
df.head(1)
```

```python
#same apply to Dep_Time column
df['Dep_hour']=df['Dep_Time'].str.split(':').str[0]

df['Dep_min']=df['Dep_Time'].str.split(':').str[1]
df.head(1)
```

```python
df.info()
```

```python
#typecast them into int

df['Dep_hour']=df['Dep_hour'].astype(int)
df['Dep_min']=df['Dep_min'].astype(int)
df.info()
```

```python
#drop the actual object column
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head(3)
```

```python
#we can take care of Route from feature Total_Stops
#so,first we will check unique and null values in Total_Stops
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'].isnull().sum()
```

```python
#let us check which value has nan stops
df[df['Total_Stops'].isnull()]
#we can replace the nan by 1 here preferably
```

```python
#let us ordinally map the categories of Total_Stops
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'2 stops':2,'1 stop':1,'3 stops':3,'4 stops':4,'nan':1})
```

```python
df.head(2)
```

```python
df.info()
```

```python
#we can drop Route now
df.drop('Route',axis=1,inplace=True)
df.head(3)
```

```python
#now let us see the Duration column , let us split it
df['Duration_hour']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
df.head(2)
```

```python
df.info()
```

```python
#typecast into int
#df['Duration_hour']=df['Duration_hour'].astype(int)
```

```python
#checking for 5m value
df[df['Duration_hour']=='5m']
```

```python
#so we will drop these 2 rows

df.drop(6474,axis=0,inplace=True)
df.drop(2660,axis=0,inplace=True)
```

```python
#and drop the Duration col
df.drop('Duration',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
#now let us look at categorical fetures like Airline
df['Airline'].unique()
```

```python
#using label encoding better for train test split
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
#and apply to the categorical variables
df['Airline']=labelencoder.fit_transform(df['Airline'])
df['Source']=labelencoder.fit_transform(df['Source'])
df['Destination']=labelencoder.fit_transform(df['Destination'])
df['Additional_Info']=labelencoder.fit_transform(df['Additional_Info'])
```
