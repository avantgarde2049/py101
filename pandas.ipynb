Dataframe merging

```python
df5 =pd.DataFrame([[3,6,4,8],[13,2,7,9],[27,13,1,0],[4,17,1,16]],columns=['X','Y','Z','Q'])
df5
```

```python
#join performedasper common column valus frm both table 
DF =pd.merge(df1,df5,right_on='Y',left_on='B')
```

```python
DF
```

```python
#using join
df1.join(df5)
```

```python
#using concat
pd.concat([df1,df5],axis=1)
```

```python
#using index merge
DF1 =pd.merge(df1,df5,right_index=True,left_index=True)
```

```python
DF1
```

```python
#converting a dataframe to numpy
DF1.to_numpy()
```

```python
#create a dictionary of objects that can be converted to series like structure
df11 = pd.DataFrame(
    {
        "A": 1.0,
        "B": pd.Timestamp("20130102"),
        "C": pd.Series(1, index=list(range(4)), dtype="float32"),
        "D": np.array([3] * 4, dtype="int32"),
        "E": pd.Categorical(["test", "train", "test", "train"]),
        "F": "foo",
        "G": pd.date_range("20130102",periods=4)    }
)
df11
```

```python
#statistic summary of a dataframe
df11.describe()
```

```python
#Grouping
d1f = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": np.random.randn(8),
        "D": np.random.randn(8),
    }
)
d1f
```

```python
d1f.groupby("A")[["C","D"]].sum()
```

```python
#slicing is similar as numpy
d1f[1:4]
```

```python
#sort by index
d1f.sort_index(axis=1,ascending=True)
```

```python
#sort by values

d1f.sort_values(by="B")
```

```python
#slicing by label location
d1f.loc[0]
```

```python
#Selecting on a multi-axis by label:

d1f.loc[:,["B","C"]]
```

```python
#selection by position
d1f.iloc[:,1:4]
```
